```python
import pandas as pd
from pandas import DataFrame
test = pd.read_csv('https://raw.githubusercontent.com/collinrijock/ml-class-collin/master/projects/test.csv', index_col="id")
train = pd.read_csv('https://github.com/collinrijock/ml-class-collin/blob/master/projects/train.csv?raw=true', index_col="id")
from sklearn.preprocessing import LabelEncoder

X_train = train.iloc[:, 0:-1].to_numpy()
Y_train = train.iloc[:, -1]
X_test = test.to_numpy()
X_train.shape
```

```python
train
```

```python
from xgboost import XGBClassifier
params = { "n_estimators": 600, 'tree_method':'gpu_hist', 'predictor':'gpu_predictor' }
model = XGBClassifier(**params)
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
predictions = model.predict(X_test)
df = pd.DataFrame(predictions)
df["id"] = test.index
df = pd.get_dummies(df, columns=[0],prefix="", prefix_sep="")
df.index = df["id"]
df
```
